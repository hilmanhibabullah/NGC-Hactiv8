Mengganti nama Colomns #, Type 1, Type 2, Total, Sp. Atk, Sp. Def menjadi Number,Type1,Type2,Total Power,Special Attack,Special Deffense. Serta mengganti data frame pokemon menjadi pokemon_new

```python
pokemon_new = pokemon.rename(columns={"#":"Number","Type 1":"Type1", "Type 2":"Type2", "Total":"Total Power","Sp. Atk":"Special Attack", "Sp. Def":"Special Deffense"})
```

Mengisi data yang kosong di colomn Type2 dengan Flying, karena Flying adalah modus di colomn Type2

```python
pokemon_new.Type2.fillna("Flying")
```

Mengisi colomn Type2 dengan fungsi yang mengisi data yang kosong di colomn Type2

```python
pokemon_new.Type2 = pokemon_new.Type2.fillna("Flying")
```

Melihat isi dan banyakanya colomn Total Power

```python
pokemon_new["Total Power"].value_counts()
```

Mengquery colomn Generation, Total Power, Special Attack, Special Deffense dengan syarat Hanya ada data Generation yang ke 6 yang memiliki total power diatas 500, special attack diatas 100, special deffense diatas 100

```python
pokemon_new[
    (pokemon_new["Generation"] == 6) &
    (pokemon_new["Total Power"] > 500 ) &
    (pokemon_new["Special Attack"] > 100) &
    (pokemon_new["Special Deffense"] > 100)
]
```

Mengquery colomn Generation, Total Power, Special Attack, Special Deffense dengan syarat Hanya ada data Generation yang ke 1 yang memiliki total power diatas 500, special attack diatas 100, special deffense diatas 100

```python
pokemon_new[
    (pokemon_new["Generation"] == 1) &
    (pokemon_new["Total Power"] > 500 ) &
    (pokemon_new["Special Attack"] > 100) &
    (pokemon_new["Special Deffense"] > 100)
]
```

Mengquery colomn Generation, Total Power, Special Attack, Special Deffense dengan syarat generasi yang lebih dari 0 yang memiliki total power diatas 500, special attack diatas 100, special deffense diatas 100

```python
pokemon_new[
(pokemon_new["Generation"] > 0) & 
    (pokemon_new["Total Power"] > 500 ) &
    (pokemon_new["Special Attack"] > 100) &
    (pokemon_new["Special Deffense"] > 100)]
```

Mengelompokkan colomn Generation dan colomn Type1 dengan menghitung jumlah dari setiap masing masing Generation (1,2,3,4,5,6)

```python
pokemon_new.groupby("Generation", sort=False)["Type1"].count()
```

Mengquery colomn Generation dan Total Power dengan syarat generasi 6 yang memiliki total power diatas atau sama dengan 400, lalu melihat dengan mengelompokkan Type1,Type2 dengan angka Total Power yang paling besar

```python
pokemon_new[
    (pokemon_new["Generation"] == 6) &
    (pokemon_new["Total Power"] >= 400 )
].groupby(["Type1", "Type2"])["Total Power"].max()
```

Mengquery colomn Generation dan Total Power dengan syarat generasi 1 yang memiliki total power diatas atau sama dengan 400, lalu melihat dengan mengelompokkan Type1,Type2 dengan angka Total Power yang paling besar

```python
pokemon_new[
    (pokemon_new["Generation"] == 1) &
    (pokemon_new["Total Power"] >= 400 )
].groupby(["Type1", "Type2"])["Total Power"].max()
```

Membuat plot bar data Generation dengan tujuan ingin melihat dan menghitung banyaknya pokemon yang di buat tiap generasi

```python
pokemon_new["Generation"].value_counts().plot(kind="bar")
```

Membuat plot dengan tujuan untuk melihat apa saja pokemon generasi 1 yang dibuat dan melihat nilai total power yang paling tinggi

```python
pokemon_new[pokemon_new["Generation"] == 1].groupby("Type1")["Total Power"].max().plot()
```

Membuat plot dengan tujuan untuk melihat apa saja pokemon generasi 6 yang dibuat dan melihat nilai total power yang paling tinggi

```python
pokemon_new[pokemon_new["Generation"] == 6].groupby("Type1")["Total Power"].max().plot()
```
